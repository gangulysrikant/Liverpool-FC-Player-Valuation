```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df_games = pd.read_csv("games.csv")
df_games
```

```python
df_players = pd.read_csv("players.csv")
df_players
```

```python
df_players['current_club_name'].unique()
```

```python
new_players_df = df_players.query('current_club_name == "Liverpool Football Club" & last_season == 2024')
new_players_df
```

```python
player_valuations_df = pd.read_csv("player_valuations.csv")
```

```python
merto_value_df = player_valuations_df[player_valuations_df['player_id'].isin(new_players_df["player_id"])]
x = merto_value_df.groupby(by=["player_id"]).max()
```

```python
merged_df = x.merge(new_players_df[['player_id', 'first_name', 'last_name']], on='player_id')
merged_df = merged_df[['first_name', 'last_name', 'player_id', 'last_season', 'market_value_in_eur']]
merged_df['market_value_in_eur'] = merged_df['market_value_in_eur'].astype('float64')
merged_df
```

```python
def format_millions(x, pos):
    'The two args are the value and tick position'
    return f'{x/1e6:.2f}M'

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))  # Set the figure size
ax = sns.barplot(data=merged_df, x=merged_df['first_name'] + ' ' + merged_df['last_name'], y='market_value_in_eur', color='skyblue')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_millions))
plt.xlabel('Player')
plt.ylabel('Max. Valuation (Millions)')
plt.title('Player Valuations [Liverpool FC 2024]')
plt.xticks(rotation=90, ha='right')  # Rotate and align x-axis labels
plt.tight_layout()
```
